## Evolutive resampling

```python
def inter(s, g):
    delta = 1 if s < g else -1
    return np.arange(s, g, delta)

def cyc(s, g):
    return conc(inter(s, g), inter(g, s))

def rep(interval, k=2):
    return np.tile(interval, k)

def conc(*args):
    return np.concatenate(args)

print(inter(0, 10))
print(inter(10, 0))
print(cyc(0, 10))
print(conc(inter(0, 10), inter(10, 0)))
print(rep(conc(inter(0, 10), inter(10, 0))))
print(conc(rep(cyc(10, 5), 3), inter(10, -2)))
```

```python
conf = build_conf("evolutive_resampling", "inet256_thick_middle", 25, 5, jump_length=5, jump_n_sample=5)
conf['manual_times'] = list(conc(inter(24, 10), rep(cyc(10, 0), 3), inter(10, -2)))
print(conf['manual_times'])
p, ani = main(conf)
```

```python
conf = build_conf("evolutive_resampling", "inet256_thick_middle_high", 25, 5, jump_length=5, jump_n_sample=5)
conf['manual_times'] = list(conc(rep(cyc(24, 14), 3), inter(24, -2)))
print(conf['manual_times'])
p, ani = main(conf)
```

### Plot results

```python
path = r"experiments/evolutive_resampling/outputs/inet256_thick_middle/its_25_jl_5_js_5/inpainted"
path_data_late = r"experiments/evolutive_resampling/outputs/inet256_thick_middle/its_25_jl_5_js_5/results/test_inet256_genhalf.csv"
path_data_early = r"experiments/evolutive_resampling/outputs/inet256_thick_middle_high/its_25_jl_5_js_5/results/test_inet256_genhalf.csv"

plt.figure(figsize=(10, 2))

for i in range(5):
    file_path = os.path.join(path, f"00000{i}.png")
    img = np.array(Image.open(file_path))
    plt.subplot(1, 5, i+1)
    plt.imshow(img)
    plt.axis('off')
plt.suptitle(f"Late resampling")
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 2))
path = r"experiments/evolutive_resampling/outputs/inet256_thick_middle_high/its_25_jl_5_js_5/inpainted"
for i in range(5):
    file_path = os.path.join(path, f"00000{i}.png")
    img = np.array(Image.open(file_path))
    plt.subplot(1, 5, i+1)
    plt.imshow(img)
    plt.axis('off')
plt.suptitle(f"Early resampling")
plt.tight_layout()
plt.show()

```

```python
df_late = pd.read_csv(path_data_late)
df_early = pd.read_csv(path_data_early)

metrics = ["lpips_vgg", "lpips_alex", "ssim"]

plt.figure(figsize=(5, 5))
labels = ['Early resampling', 'Late resampling']
for i, metric in enumerate(metrics):
    avg_early, std_early = df_early[metric].mean(), df_early[metric].std()
    avg_late, std_late = df_late[metric].mean(), df_late[metric].std()

    if metric in ['ssim']:
        avg_early, avg_late = 1-avg_early, 1-avg_late

    plt.errorbar([i], [avg_early], [std_early], label='Early', color='blue', marker='+', capsize=5)
    plt.errorbar([i], [avg_late], [std_late], label='Late', color='orange', marker='x', capsize=5)


metrics[2] = "1-ssim"
plt.xticks([0, 1, 2], metrics)


handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

plt.tight_layout()
plt.show()
```
